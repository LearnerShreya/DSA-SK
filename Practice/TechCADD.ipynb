````python
a = [1, 3, 4, 5, 0, 2]
largest = float('-inf')
second_largest = float('-inf')

for x in a:
    if x > largest:
        second_largest = largest
        largest = x
    elif largest > x > second_largest:
        second_largest = x

print(second_largest)
print(largest)
````
